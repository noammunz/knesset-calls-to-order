```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Windows Results

```python
results_path = '/home/nogaschw/Call-to-order/Data2/windowed_results_new.csv'
results = pd.read_csv(results_path)
results = results.sort_values(by='window_size', ascending=True)
# round to 3 decimal places
results = results.round(3)
results = results[['window_size', 'precision', 'recall', 'f1', 'auc']]
results
```

```python
a = results['f1'].to_list()
b = list(range(1, 11))
# line plot
title = 'F1 score for different window sizes'
xlabel = 'Window size'
ylabel = 'F1 score'
x = b
y = a
# plot
fig, ax = plt.subplots()
ax.plot(x, y, marker='o', color='teal')
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
# x ticks fgrom 1 to 10
ax.set_xticks(b)
# ylim 0 to 0.8
ax.set_ylim(0.3, 0.7)
plt.show()
```
